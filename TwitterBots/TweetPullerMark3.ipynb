```python
#Construct a dataframe to hold all of the tweets
diction = {'text':[]}
db = pd.DataFrame(diction)

#Loop through all accounts
countPerson = 0
for person in people:
    print("Person " + str(countPerson))
    try:
        #Get past 200 tweets
        userTweets = api.user_timeline(screen_name=person.screen_name,
                                   count=200,
                                   include_rts = False,
                                   tweet_mode = 'extended'
                                   )
        #Remove replies
        for tweet in userTweets:
            if tweet.in_reply_to_status_id is None:
                db.loc[len(db.index)] = tweet.full_text
                
        print("Successfully retrieved tweets from this account")
        
    except:
        print("Failed to retrieve tweets from this account")
        
    countPerson += 1
    
#Save to CSV
db.to_csv(CSV_FILE_PATH, mode='a', index=False, header=False)
print("Successfully Saved to CSV")
```
